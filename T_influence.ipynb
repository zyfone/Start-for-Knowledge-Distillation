```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format="retina"
```

探索温度T对logits的影响

```python
logits=np.array([-10,-5,1,2,7,11])
```

```python
softmax_one=np.exp(logits)/sum(np.exp(logits))
softmax5=np.exp(logits/5)/sum(np.exp(logits/5))
softmax10=np.exp(logits/10)/sum(np.exp(logits/10))
softmax100=np.exp(logits/100)/sum(np.exp(logits/100))
```

```python
softmax_one
```

```python
plt.plot(softmax_one,label="t=1")
plt.plot(softmax5,label="t=5")
plt.plot(softmax10,label="t=10")
plt.plot(softmax100,label="t=100")
plt.xticks(np.arange(5),['class1','class2','class3','class4','class5'])
plt.legend()
plt.show()
```
